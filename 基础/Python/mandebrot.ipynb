```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def iter_point(z0):
    z = z0
    for i in np.arange(1, 256):
        if np.abs(z) > 4:
            break
        z = z*z + z0
    return i
```

```python
def mandelbrot(cx, cy, d, n=512):
    x0, x1, y0, y1 = cx-d, cx+d, cy-d, cy+d
    y, x = np.ogrid[y0:y1:n*1j, x0:x1:n*1j]
    c = x + y*1j
    return np.frompyfunc(iter_point,1,1)(c).astype(np.float)
```

```python
def draw_m(cx, cy, d, n=200):
    plt.imshow(mandelbrot(cx, cy, d,n), cmap=cm.Blues_r)
    plt.gca().set_axis_off()
```

```python
x, y = 0.27, 0.6
```

```python
plt.figure(figsize=(9, 6))
plt.subplot(231)
draw_m(-0.5, 0, 1.5)
for i in np.arange(2, 7):
    plt.subplot(230+i)
    draw_m(x, y, 0.2**(i-1))
plt.subplots_adjust(0, 0, 1, 1, 0.0, 0)
```
